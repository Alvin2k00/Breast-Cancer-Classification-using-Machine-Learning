```python
X=df2.iloc[:, 1:]
y=df2.iloc[:, 0]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train.info()
```

```python
X_test.info()
```

# **Feature Scaling**

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

# **IMPLEMENTING ALGORITHMS**

# **KNN Algorithm**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)

knn_test=classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,knn_test)
plt.rcParams["figure.figsize"] = [8, 6]
sns.heatmap(cm,annot=True,cmap="PiYG")
plt.ylabel('Predicted')
plt.xlabel('Actual')
```

```python
print(classification_report(y_test, knn_test))
```

```python
print((accuracy_score(y_test, knn_test))*100)
```

# **Decision Tree**

```python
from sklearn import tree,metrics
usingtree=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=0)
usingtree.fit(X_train, y_train)
```

```python
decision_tree_test=usingtree.predict(X_test)
incorrect_classification=(y_test!=decision_tree_test).sum()
print(f"Samples which were not correctly classified: {incorrect_classification}")
accuracy = metrics.accuracy_score(y_test, decision_tree_test)
print(f"Accuracy using Decision Trees: {accuracy*100}")
```

```python
print(classification_report(y_test, decision_tree_test))
```

```python
cm=confusion_matrix(y_test,decision_tree_test)
plt.rcParams["figure.figsize"] = [8, 6]
sns.heatmap(cm,annot=True,cmap="BuPu")
plt.title('Confusion Matrix for Decision tree')
plt.ylabel('Predicted')
plt.xlabel('Actual')
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
forest_model = RandomForestClassifier(n_estimators=100)
forest_model.fit(X_train, y_train)

# Making predictions
random_forest_test = forest_model.predict(X_test)

# Getting accuracy and report
accuracy = accuracy_score(y_test, random_forest_test)
report = classification_report(y_test, random_forest_test)

print(f"Model Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```

```python
cm=confusion_matrix(y_test,random_forest_test)
plt.rcParams["figure.figsize"] = [8, 6]
sns.heatmap(cm,annot=True,cmap="Greens")
plt.title('Confusion Matrix for Random Forest')
plt.ylabel('Predicted')
plt.xlabel('Actual')
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
regression_model = LogisticRegression(random_state=0)
regression_model.fit(X_train, y_train)

# Making predictions
logistic_regression_test = regression_model.predict(X_test)

print('Model Accuracy:')
print(accuracy_score(y_test, logistic_regression_test))
print('Classification Report: ')
print(classification_report(y_test, logistic_regression_test))
```

```python
cm=confusion_matrix(y_test,logistic_regression_test)
plt.rcParams["figure.figsize"] = [8, 6]
sns.heatmap(cm,annot=True,cmap="Blues")
plt.title('Confusion Matrix for Logistic Regression')
plt.ylabel('Predicted')
plt.xlabel('Actual')
```

# **SVM (Support Vector Machine)**

```python
from sklearn.svm import SVC
model_svm = SVC(random_state=0)
model_svm.fit(X_train, y_train)

# Making predictions
svm_test = model_svm.predict(X_test)

print('Model Accuracy:')
print(accuracy_score(y_test, svm_test))
print('Classification Report: ')
print(classification_report(y_test, svm_test))
```

```python
cm=confusion_matrix(y_test,svm_test)
plt.rcParams["figure.figsize"] = [8, 6]
sns.heatmap(cm,annot=True,cmap="YlGnBu")
plt.title('Confusion Matrix for SVM')
plt.ylabel('Predicted')
plt.xlabel('Actual')
```

# **Naive Bayes (NB) Algorithm**

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Making predictions
gaussiannb_test = nb_model.predict(X_test)

print('Model Accuracy:')
print(accuracy_score(y_test, gaussiannb_test))
print('Classification Report: ')
print(classification_report(y_test, gaussiannb_test))
```
